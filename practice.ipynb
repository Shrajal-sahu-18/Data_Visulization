```python
from matplotlib import pyplot as plt

models = ["Laptop A", "Laptop B", "Laptop C", "Laptop D", "Laptop E"]
ram = [8, 16, 8, 32, 16]
price = [45000, 65000, 52000, 98000, 72000]

plt.figure(figsize=(7,5))
plt.scatter(ram, price)
for x, y, label in zip(ram, price, models):
    plt.annotate(label, (x, y))

plt.xlabel("RAM (GB)")
plt.ylabel("Price (₹)")
plt.title("Laptop Data Visualization: RAM vs Price")
plt.tight_layout()
```
