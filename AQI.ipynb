```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create dataset
aqi_data = pd.DataFrame({
    'Month': ['Jan','Feb','Mar','Apr','May','Jun',
              'Jul','Aug','Sep','Oct','Nov','Dec'],
    'AQI': [180,165,140,120,95,85,75,80,110,160,220,250]
})
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=aqi_data, x='Month', y='AQI', marker='o')

plt.title('Monthly Air Quality Index (AQI)')
plt.xlabel('Month')
plt.ylabel('AQI')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=aqi_data, x='Month', y='AQI')
plt.title('Monthly AQI Comparison')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.fill_between(aqi_data['Month'], aqi_data['AQI'], alpha=0.5)
plt.plot(aqi_data['Month'], aqi_data['AQI'], marker='o')
plt.title('AQI Area Chart')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=aqi_data['AQI'])
plt.title('AQI Box Plot')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.stripplot(y=aqi_data['AQI'], size=8)
plt.title('AQI Strip Plot')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.swarmplot(y=aqi_data['AQI'], size=8)
plt.title('AQI Swarm Plot')
plt.show()
```
