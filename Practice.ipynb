```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("WorldCupMatches.csv")
```

```python
df.head()
```

```python
print(df.info())
```

```python
df = df.dropna()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(14,6))
sns.countplot(x='Year', data=df)
plt.xticks(rotation=90)
plt.title("Number of Matches Per Year")
plt.show()
```

```python
plt.figure(figsize=(12,6))

top_teams = df['Home Team Name'].value_counts().head(10)

sns.barplot(
    x=top_teams.index,
    y=top_teams.values,
    palette='viridis'
)

plt.title("Top 10 Home Teams")
plt.xlabel("Team")
plt.ylabel("Matches")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df['Attendance'],
    bins=30,
    kde=True,
    color='blue'
)

plt.title("Attendance Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.kdeplot(
    df['Home Team Goals'],
    shade=True,
    color='red'
)

plt.title("Density Plot of Home Team Goals")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Home Team Goals',
    y='Away Team Goals',
    hue='Year',
    data=df
)

plt.title("Home Goals vs Away Goals")
plt.show()
```

```python
goals_per_year = df.groupby('Year')['Home Team Goals'].sum().reset_index()

plt.figure(figsize=(12,6))

sns.lineplot(
    x='Year',
    y='Home Team Goals',
    data=goals_per_year,
    marker='o'
)

plt.title("Total Home Goals Per Year")
plt.show()
```

```python
plt.figure(figsize=(14,6))

sns.boxplot(
    x='Stage',
    y='Attendance',
    data=df
)

plt.xticks(rotation=90)
plt.title("Attendance by Match Stage")
plt.show()
```

```python
plt.figure(figsize=(14,6))

sns.violinplot(
    x='Stage',
    y='Home Team Goals',
    data=df
)

plt.xticks(rotation=90)
plt.title("Goal Distribution by Stage")
plt.show()
```

```python
sns.pairplot(df[
    ['Home Team Goals',
     'Away Team Goals',
     'Attendance']
])

plt.show()
```

```python
plt.figure(figsize=(8,6))

corr = df[
    ['Home Team Goals',
     'Away Team Goals',
     'Attendance']
].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.jointplot(
    x='Home Team Goals',
    y='Away Team Goals',
    data=df,
    kind='scatter'
)

plt.show()
```

```python
plt.figure(figsize=(14,6))

sns.stripplot(
    x='Stage',
    y='Home Team Goals',
    data=df
)

plt.xticks(rotation=90)
plt.title("Strip Plot")
plt.show()
```

```python
plt.figure(figsize=(14,6))

sns.swarmplot(
    x='Stage',
    y='Away Team Goals',
    data=df
)

plt.xticks(rotation=90)
plt.title("Swarm Plot")
plt.show()
```
