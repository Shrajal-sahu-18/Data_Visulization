```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Student_Marks.csv")
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="time_study", y="Marks")
plt.title("Study Time vs Marks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Marks"], bins=10, kde=True)
plt.title("Distribution of Marks")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=df["Marks"])
plt.title("Box Plot of Marks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="number_courses", y="Marks")
plt.title("Number of Courses vs Marks")
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.regplot(
    data=df,
    x="time_study",
    y="Marks"
)
plt.title("Study Time vs Marks with Trend Line")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["time_study"], bins=10, kde=True)
plt.title("Study Time Distribution")
plt.show()
```
